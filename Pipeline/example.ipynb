```python
#access the generated balloons
balloons_folder = 'generate_output_balloons_unit/speech_balloons'
balloon_coordinates_folder = 'generate_output_balloons_unit/speech_balloons_coordinates'
detection_folder = 'generate_output_balloons_unit/detections'


ballons = list(Path(balloons_folder).rglob('*.png'))
balloon_coordinates = list(Path(balloon_coordinates_folder).rglob('*.txt'))
detections = list(Path(detection_folder).rglob('*.txt'))

#read the detections file
with open(detections[0], 'r') as f:
    detections = f.read().splitlines()
    #split detections by spaces
    detections = [x.split() for x in detections]
    #sort detections by the 4th and 5th element (x and y coordinates)

    detections.sort(key = lambda x: (float(x[4]), float(x[3])))
    #detections_type = [x[0] for x in detections]
    detections_coordinates = [(x[3], x[4]) for x in detections]
print(detections_coordinates)

balloon_sorted = []
#loop ballons and append to sorted list based on coordinates
for i in range(len(balloon_coordinates)):
    #read balloon coordinates[i].txt and save in a 4 element list
    with open(balloon_coordinates[i], 'r') as f:
        coordinates = f.read().split()
    #append the balloon image to the sorted list
    balloon_sorted.append((coordinates, ballons[i]))

#sort the list based on the first 2 elements of coordinates (float)
balloon_sorted.sort(key = lambda x: (float(x[0][1]), float(x[0][0])))

#imshow the sorted balloons 
for i in range(len(balloon_sorted)):
    
    print(balloon_sorted[i][0])
    print(detections_coordinates[i])
    plt.imshow(plt.imread(balloon_sorted[i][1]))
    plt.show()
```
